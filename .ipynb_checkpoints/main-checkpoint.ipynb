```python
#Importong libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.decomposition import TruncatedSVD
```

```python
#Reading the data
data = pd.read_csv("spam.csv", encoding = 'latin-1')
```

```python
'''
 As we see there is no important data to be used, so we will delete the columns.
'''
```

```python
#Checking if the deletion prooccess is what we need
data.drop(columns = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"])
```

```python
#As it is the way we want the data to be deleted, now we actually delete the data
data.drop(columns = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], inplace = True)
```

```python
#Checking if the deletion was successful
data
```

```python
#Converting text data into numbers: ham -> 1, spam -> -1
data.v1 = data.v1.map({"ham": 1, "spam": -1})
```

```python
#Checking if the convertion was successful
data
```

```python
#Splitting data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(data.v2, data.v1, test_size = 0.2, random_state = 1, stratify = data.v1)
```

```python
tfidf = TfidfVectorizer(stop_words = "english")
```

```python
X_train_tfidf = tfidf.fit_transform(X_train)
```

```python
X_test_tfidf = tfidf.transform(X_test)
```

```python
#Initializing SVM
sv = SVC(C = 3.0, kernel = 'linear', gamma = 1.0)
```

```python
#Fitting the model
sv.fit(X_train_tfidf, y_train)
```

```python
#Predicting and checking the accuracy score
y_pred = sv.predict(X_test_tfidf)
print(f"The prediction accuracy of the model using the testing dataset is: {100 * accuracy_score(y_pred, y_test):.3f}")
```
